```python
normalized = raw.apply(lambda col : col - background['background'])
normalized.head()
```

```python
normalized['frame'] = normalized.index
normalized
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
sns.clustermap(
    normalized.T,
    cmap = 'OrRd',
    col_cluster=False
)
```

```python


import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
sns.heatmap(
    normalized.iloc[520-1:600-1].T,
    cmap = 'OrRd',
)
```

```python
normalized = normalized.melt(var_name = 'cell', value_name = 'fluorescence', id_vars=['frame'])
normalized
```

```python
normalized = normalized[['cell', 'frame', 'fluorescence']]
```

```python
normalized.to_csv('normazlized_tidy.csv', index=False)
```

```python
baseline_mean = normalized.groupby('cell').apply(lambda x : x.iloc[519:599]['fluorescence'].mean()).to_dict()
baseline_sd = normalized.groupby('cell').apply(lambda x : x.iloc[519:599]['fluorescence'].std()).to_dict()
```

```python
normalized['baseline_mean'] = normalized['cell'].apply(lambda x : baseline_mean[x])
normalized['baseline_sd'] = normalized['cell'].apply(lambda x : baseline_sd[x])
```

```python
normalized['amplitude'] = (normalized['fluorescence'] - normalized['baseline_mean']) / normalized['baseline_mean']
```

```python
normalized_summary = normalized.groupby(['frame', 'cell'])['amplitude'].first().unstack()
```

```python
###### import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
sns.heatmap(
    normalized_summary.T.iloc[160:180],
    #cmap = 'OrRd',
    vmax=3,
    vmin=-3,
    cmap='RdBu_r'
)
```

```python
my_cells = normalized[normalized['cell'].isin([216, 175, 168])]
```

```python
my_cells['cell_name'] = 'c' + my_cells['cell'].astype(str)
my_cells
```

```python
sns.relplot(
    data = my_cells,
    kind='line',
    x = 'frame',
    y = 'amplitude',
    hue = 'cell_name',
    aspect =3
)
```

```python
responders = normalized[normalized['amplitude'] > 3 * normalized['baseline_sd']]
```

```python
def classify(frame_no):
    if 700 < frame_no < 840:
        return 'first 50g'
    elif 9700 < frame_no < 9829:
        return 'second 50g'
    else:
        return 'nothing'

normalized['event'] = normalized['frame'].apply(classify)
normalized
```

```python
my_events = normalized[normalized['event'] != 'nothing']
```

```python
my_events['event'].value_counts()
```

```python
summary = my_events.groupby(['cell', 'event'], as_index=False)['amplitude'].mean()
summary
```

```python
sns.catplot(
    data = summary,
    x = 'event',
    y = 'amplitude',
    kind = 'swarm',
    aspect = 5
)
```
